```python
def mima():
    a=input("输入密码")
    print("长度为",len(a))
    if a.isalpha()==True:
        print("没有数字")
    if a.isdigit()==True:
        print("没有字母")
    else:
        print("有字母也有字符")
```

```python
mima()
```

```python
def mima1():
    n1=0 #大写
    n2=0 #小写
    n3=0 #数字
    a=input("输入密码")
    for i in a:
        if i.isalpha():
            n2+=1
        if i.isdigit():
            n3+=1
        if i.isupper():
            n1+=1
    if n1==0:
        print("没有大写")
    elif n2==0:
        print("没有小写")
    elif n3==0:
        print("没有数字")
    else:
        print("密码设置完成")
```

```python
mima1()
```
